```python
coffee={"Espresso":1.50,"Americano":2.00,"Caffe latte":2.25,"Mocha":2.25,"Cappucino":2.25,"Macchiato":2.25}
tea={"White tea":1.5,"Green tea":2.00,"Oolong tea":2.25,"Earl Grey":2.25,"Matcha":2.25}
pastries={"Croissant":5.50,"Cinnamon roll":4.50,"Banana bread":4.25,"Pumpkin bread":3.75,"Sticky bun":4.25,"Apple turnover":3.75}

print("Welcome to MC Cafe!")
q1=input("Do you wnat to see the menu? (yes/no)" )
if q1=="yes":
#while True:
    print("Menu")
    print("Coffee")
    for x in coffee:
        print(f"{x}- ${coffee[x]}")    

    print("Tea")
    for x in tea:
        print(f"{x} - ${tea[x]}")

    print("Pastries")
    for x in pastries:
        print(f"{x} - ${pastries[x]}")

q2=input("Are you ready to order? (yes/no)")
customer_order={}
while True:
    initem=input("Enter your an item: ")
    inq=int(input("Enter your quantity: "))
    customer_order.update({initem : inq})
    print(customer_order)
    q3=input("Do you want order another item? (yes/no) ")
    if q3=="yes":
        continue
    else:
        print(customer_order)
        break

  
sub=0.0
for x in customer_order:

    if x in coffee:
        y=coffee.get(x)

        z=customer_order.get(x)

        sub+=y*z
#        print(sub)
    if x in tea:
        y=tea.get(x)

        z=customer_order.get(x)

        sub+=y*z
#        print(sub)
    if x in pastries:
        y=pastries.get(x)

        z=customer_order.get(x)

        sub+=y*z
#        print(sub)
total=sub+sub*0.06
print(f"Total is ${total}")
print("Thank you for visiting MC Cafe!")
      
          
```
